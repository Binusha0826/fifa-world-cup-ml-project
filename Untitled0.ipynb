```python
import pandas as pd
df = pd.read_csv('FIFA_World_Cup_Champions_1930_2022.csv')
```

```python
#rows and column count
print(df.shape)
```

```python
#get first five rows
df.head()
```

```python
print(df.columns.tolist())
```

```python
#to get null count and data type
df.info()
```

```python
#to get missing value
df.isnull().sum()
```

```python
#drop highst no of missing values columns
df_clean = df.drop(columns=['golden_ball', 'best_young_player', 'golden_glove', 'mascot'])
```

```python
#select features and targets
features = ['host_country', 'total_teams', 'matches_played',
            'attendance', 'continent_winner', 'winner_titles_after_win']
target = 'win_method'
```

```python
#input data
X = df_clean[features]
```

```python
#predict column
y = df_clean[target]
```

```python
#target unique values
print(y.value_counts())
```

 Regular Time (14), Extra Time (5), Penalties (3) . This is imbalanced dataset, because regular time is very high

```python
# Categorical columns encoding (text to num)
X_encoded = pd.get_dummies(X, columns=['host_country', 'continent_winner'])
```

```python
#encoding labels for target variales
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
#check encode
print(X_encoded.head())
print(y_encoded)
print(le.classes_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y_encoded, test_size=0.2, random_state=42
)
```

```python
print("Train size:", X_train.shape)
print("Test size:", X_test.shape)
```

Train: 17 rows, Test: 5 rows, So dataset is small

```python
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree model
model = DecisionTreeClassifier(random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
print("Model trained")
```

```python
#predict test data
y_pred = model.predict(X_test)

print("Actual values:   ", y_test)
print("Predicted values:", y_pred)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Calculate Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Detailed report (precision, recall, f1-score)
print(classification_report(y_test, y_pred, zero_division=0))
```

Why did this happen?
Too few rows (only 22)
Small test set (only 5 rows)
Class imbalance
Features don't have a strong signal

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

model2 = RandomForestClassifier(random_state=42)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)

accuracy2 = accuracy_score(y_test, y_pred2)
print("Random Forest Accuracy:", accuracy2)
```
